## Imports

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

### Import data pickle Llava

```python
with open("pickle_data/results_semantic_sim_expt_cot_llava.pkl", "rb") as f:
    results_semantic_sim_expt_cot_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_semantic_sim_answer_original_llava.pkl", "rb") as f:
    results_semantic_sim_answer_original_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_semantic_sim_answer_optimized_llava.pkl", "rb") as f:
    results_semantic_sim_answer_optimized_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_bleu_expt_cot_llava.pkl", "rb") as f:
    results_bleu_expt_cot_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_bleu_answer_original_llava.pkl", "rb") as f:
    results_bleu_answer_original_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_bleu_answer_optimized_llava.pkl", "rb") as f:
    results_bleu_answer_optimized_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_rouge_expt_cot_llava.pkl", "rb") as f:
    results_rouge_expt_cot_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_rouge_answer_original_llava.pkl", "rb") as f:
    results_rouge_answer_original_llava=loaded_results = pickle.load(f)

with open("pickle_data/results_rouge_answer_optimized_llava.pkl", "rb") as f:
    results_rouge_answer_optimized_llava=loaded_results = pickle.load(f)
```

## Grafici Llava

#### Line Chart Answer Expert Prompting and CoT

```python
labels = ['Answer1', 'Answer2', 'Answer3', 'Answer4', 'Answer5','Answer6','Answer7','Answer8','Answer9','Answer10','Answer11','Answer12','Answer13','Answer14','Answer15','Answer16','Answer17','Answer18']

# Line Chart
plt.figure(figsize=(19, 10))
plt.plot(labels, results_semantic_sim_expt_cot_llava, marker='o', label="Semantic Similarity")
plt.plot(labels, results_bleu_expt_cot_llava, marker='s', label="BLEU")
plt.plot(labels, results_rouge_expt_cot_llava, marker='^', label="ROUGE")
plt.xlabel("Answers")
plt.ylabel("Score")
plt.title("Andamento delle Metriche Semantic Similarity, BLEU, ROUGE")
plt.legend()
plt.grid(True)
plt.yticks(np.arange(0.0, 1.0, 0.05))
plt.show()
```
